```python
import pandas as pd
import simplejson
import copy
```

```python
df = pd.read_csv('../input/new_nin_2.csv')
df
```

```python
df.columns
```

```python
json_template = {
    "nin_code":"",
    "ingredient_name":"",
    "ingredient_description":"[]",
    "macros":{"protein":"","total fat":"","fiber":"","carbs":"","energy":""},
    "micros":{"calcium":"0","iron":"0","manganese":"0","potassium":"0","phosphorus":"0","sodium":"0","zinc":"0","fructose":"0","glucose":"0","sucrose":"0","starch":"0","magnesium":"0","cholestrol":"0","water":"0","folic acid":"0"}
}

json_list = []
for index, row in df.iterrows():
    if row['ingredient_name'] != row['ingredient_name']:
        continue
    ingredient_name = row['ingredient_name']
    energy = str(row['Total Calories'])
    fat = str(row['Fat'])
    carbs = str(row['Carbs'])
    fiber = str(row['Fiber'])
    protein = str(row['Protein'])
    
    instance = json_template
    instance['ingredient_name'] = ingredient_name
    instance['macros']['protein'] = protein
    instance['macros']['total fat'] = fat
    instance['macros']['fiber'] = fiber
    instance['macros']['carbs'] = carbs
    instance['macros']['energy'] = energy
    json_list.append(copy.deepcopy(instance))
```

```python
with open('../output/new_nin.json', 'w', encoding='utf-8') as f:
    simplejson.dump(json_list, f, indent=4, ensure_ascii=True)
```
