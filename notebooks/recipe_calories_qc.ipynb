```python
import pandas as pd
import json
import requests
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Create a sheet with all recipes along with with serving weight and calories

```python
df = pd.DataFrame()
payload={}
headers = {}

# mention the total number of recipes
num_of_recipes = 338

for recipe_id in range(num_of_recipes+1):
    url = f"http://18.223.178.105:5000/api/v1/recipe_management/recipe/{recipe_id}"     # dev
    # url = f"http://13.234.6.215:5000/api/v1/recipe_management/recipe/{recipe_id}"     # prod
    response = requests.request("GET", url, headers=headers, data=payload)
    response = json.loads(response.text)

    try:
        if 'message' in response.keys():
            print(f'recipe {recipe_id} error')
            df = df.append({
                'recipe_id': recipe_id,
                'recipe_name': None,
                'serving': None,
                'per_serving_wt': None,
                # 'calories_all_servings': None,
                'calories_per_serving': None,
            },
            ignore_index=True
            )
            continue

        recipe = response['recipe']
        recipe_name = recipe['recipe_name']
        per_serving_wt = recipe['per_serving']
        serving = recipe['serving']
        macros = recipe['macros']
        energy = [mac['value'] for mac in macros if mac['key']== 'energy'][0]

        
        df = df.append({
            'recipe_id': recipe_id,
            'recipe_name': recipe_name,
            'serving': serving,
            'per_serving_wt': round(per_serving_wt, 2),
            # 'calories_all_servings': energy,
            'calories_per_serving': round((energy / serving), 2),
        },
        ignore_index=True
        )
        print(f'recipe {recipe_id} done')
    
    except Exception as e:
        print('Error: ', e)
```

```python
df.to_csv('../output/Calories of meals(15Nov).csv', index=False)
```
